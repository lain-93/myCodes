```python
points = np.arange(-5, 5 , 0.01)
```

```python
dx , dy = np.meshgrid(points , points)
```

```python
dx
```

```python
dy
```

```python
z = (np.sin(dx) + np.sin(dy))
```

```python
z
```

```python
plt.imshow(z)
```

```python
plt.imshow(z)
plt.colorbar()
plt.title('this is the image of sin(x)+sin(y)')
```
