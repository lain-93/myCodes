```python
import numpy as np
from pandas import Series, DataFrame 
import pandas as pd
```

```python
dframe1 = DataFrame({'key':['x','y','z','x','x'], 'dataset1':np.arange(5)})
dframe1
```

```python
dframe2 = DataFrame({'key':['y','z','q','z'], 'dataset2':[1,2,3,4]})
dframe2
```

```python
pd.merge(dframe1,dframe2)
```

```python
df_left = DataFrame({'key1':['SF','SF','LA'], 'key2': ['one','two', 'one'], 'left_data':[10,20,30] })
```

```python
df_right = DataFrame({'key1':['SF','SF','LA','LA'], 'key2': ['one','one', 'one', 'two'], 'right_data':[40,50,60,70] })
```

```python
df_left
```

```python
df_right
```

```python
pd.merge(df_right, df_left, on=['key1','key2'], how='outer')
```

```python
pd.merge(df_right, df_left, on=['key1'])
```

```python
pd.merge(df_right, df_left, on=['key1'], suffixes=('_righty', '_lefty'))
```
